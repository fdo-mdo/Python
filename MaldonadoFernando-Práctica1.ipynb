### Tu trabajo consiste en analizar las palabras que los autores han usado en cada obra.

Para ello ejecuta el siguiente codigo para obtener las palabras que se usaron en el primer libro.

```python
libro = open("./libros/10.txt", "r") #Aura, Carlos Fuentes
lista = []
for linea in libro:
    stripped_line = linea.strip()
    line_list = stripped_line.split()
    if line_list != []:
        for i in range(len(line_list)):
            line_list[i] = line_list[i].lower()
        lista.append(line_list)
libro.close()
lista = sum(lista) #lista de palabras usadas en el libro 1
```

## Primera Parte: Ordenando los datos

Nuestra base de datos por el momento, solo tiene informacion de las palabras que hay en el texto.  Para estructurar los datos realizaremos los siguientes pasos:

1- Obtendremos las palabras unicas que conforman el texto.

2- Contaremos cuantas veces aparece cada palabra. Con ello tendremos un nuevo dato, la frecuencia de aparicion.

3- Ordenaremos las palabras de acuerdo a su frecuencia. Donde el primer lugar  correspondera a la palabra mas frecuente, el segundo lugar a la segunda mas frecuente y asi sucesivamente con todas las demas.


Estos tres pasos se pueden hacer con listas, condicionales y bucles,  pero seria muy tardado el computo.  Usemos mejor diccionarios.

```python
dict0={palabra:lista.count(palabra) for palabra in lista}
```

```python
dict0_orden = sorted(dict0.items(), key= lambda x: x[1],reverse=True)
```

## Segunda Parte: Analisis de las frecuencias

La estructura de la base de datos consiste en tener las palabras ordenadas por rangos $k$, donde la frecuencia de cada palabra $f(k)$ esta asociada a un rango.

Con lo anterior se puede obtener informacion del como se distribuyen las palabras en un texto de acuerdo a su frecuencia. Para ello se suguiere lo siguiente:

1- Normalizar las frecuencias. Divide la frecuencia de cada palabra entre el valor mas grnde de estas.

2- Haz la grafica $k$ vs $f(k)$.  Usa escala logaritmica en ambos ejes.

3- ¿que observas de la grafica anterior?. Para obtener informacion adicional puedes observar los primeros 10  valores de las frecuencias normalizadas.

4- A partir de lo que observas en la grafica, obten una expresion analitica que describa la curva y graficala en el mismo espacio que la anterior.  Para la expresion analitica, no es necesario que hagas regresiones, basta con recordar la grafica de las funciones mas conocidas.

5- Dale presentacion a tu grafica (titulo, nombre de los ejes, labels, color de la grafica, tipo de marker, etc).

```python
pal, frec=[],[]
for i in range(0, len(dict0_orden)):
    palabra = dict0_orden[i][0]
    frecuencia = dict0_orden[i][1]
    pal.append(palabra)
    frec.append(frecuencia)
    
frec=array(frec)
frec_norm=frec/max(frec)
```

## Trabajando con las palabras

Ya hemos trabajado con las frecuencias de las palabras,  pero tambien podemos obtener informacion sobre las palabras mismas.

Recordemos que las listas de palabras estan ordenadas de acuerdo a su frecuencia, donde cada palabra ocupa un lugar especifico en el ordenamiento.  A este lugar especifico lo llamaremos rango $k$ .

El proximo resultado a encontrar parte del algoritmo de $ \textbf{Diversidad de Rango d(k)} $. 

Para obtener la diversidad de rango pensemos disponemos de 50 listas de palabras (todas ellas cumplen el ordenamiento previo).

1- Tomamos el primer rango $k=1$ de todas las 50 listas, y contamos cuantas palabras diferentes ocupan ese mismo rango.

2- A la cantidad anterior la dividiremos entre 50 (para tener cantidades normalizadas) obteniendo asi la diversidad para el primer rango $d(1)$.  Este sera el primer punto de nuestra grafica $1$ vs $d(1)$.

3- Repetimos los pasos 1 y 2 para todos los rangos en la lista.



## Diversidad de Rango

El siguiente archivo contiene las 2500 palabras mas usadas de los 50 libros anteriores.  Este archivo esta estructurado en una lista de listas, donde Palabras[0] corresponde a las palabras mas usadas en el libro 1, Palabras[1] corresponde a las palabras mas usadas en el libro 2,   y asi sucesivamente.

1- Construye un algoritmo para encontrar la diversidad de rango del listado anterior. Te sugerimos primero encontrar el valor de diversidad de cada rango, y al final graficar los rangos vs su diversidad.

Para la grafica usa markers redondos (como un scatter plot) y usa escala logaritmica en el eje x.


```python
Palabras = load("libros/Palabras.npy").tolist()
```

```python
len(Palabras[8])
```

```python
drango=[]
for i in range(2500):
    contador=set()
    for j in range(50):
        contador.add(Palabras[j][i])
    drango.append(len(contador))
print(drango)
drangonorm=[]
for i in range(2500):
    drangonorm.append(float(drango[i])/50)
print(drangonorm)
```

```python
rang=arange(1,2501,1)
plot(rang,drangonorm,".",color="b")
xscale('log')
title('Distribución de la diversidad de rangos')
xlabel('Rangos')
ylabel('Diversidad de rangos')
plt.grid(True)
```

Con esto podemos darnos cuenta que en todos los libros analizados, tal como dice nuestra intuición, las palabras menos usadas de un libro no suelen coincidir con las de otro, por lo que el valor de la diversidad de rango tiende a 1 mientras menos comunes o 'exóticas' sean las palabras usadas por el autor. De la misma manera, las palabras más usadas tienden a usarse entre la mayoría de los autores por la misma naturaleza del lenguaje.

$ \large{\textbf{EXTRA1}} $

La curva de diversidad puede ser descrita por la siguiente expresion (dependiente de $\mu$ y $\sigma$).

$$d(k) = \frac{1}{\sigma\sqrt{2\pi}}\int_{-\infty }^{\log_{10^{k}}} e^{-\frac{\left ( y-\mu \right )^{2}}{2\sigma^{2}}}dy.$$ 	


La siguiente funcion devuelve los rangos y sus evaluaciones en la integral anterior.  

A partir de los valores que obtuviste de la diversidad, evalualos en la funcion curva_ms para realizar su ajuste.

Prueba con valores de $\mu = [1, 2]$ y $\sigma = [0.1, 2.1]$. Encuentra la pareja de mu y sigma que mejor ajuste.



$ \large{\textbf{EXTRA2}} $

Encuentra la fiabilidad del ajuste con el metodo que prefieras.

```python
from scipy import integrate

def curva_ms(diversidad, mu, sigma):
    K = arange(0,len(diversidad))+1
    val = []
    def fsig(x):
        c = 1 /(sigma*sqrt(2*pi)) 
        y = exp( (-(x-mu)**2) / (2* sigma**2) ) 
        return y*c
    for i in range(0,len(K)):
        k = K[i]
        if log10(k) == 0:
            b = log10(2)/2
        elif log10(k) >0:
            b = log10(k)
        res = integrate.quad(fsig, -inf, b)
        val.append(res[0])
    return K, val
```

Usaremos el error cuadratico medio del método de mínimos cuadrados para encontrar los valores de mu y sigma para los cuales el error sea mínimo, es decir, que mejor ajusten la curva.

```python
from math import sqrt
def errorcuad(diversidad,datos_ajuste): #Error cuadratico medio
    err=0
    for i in range(len(diversidad)):
        err=err+(diversidad[i]-datos_ajuste[i])**2
    errc=sqrt(err/len(diversidad))
    return errc
```

```python
listaerr=[]
listamusig=[]
for i in arange(1,2,0.1):
    for j in arange(0.1,2.1,0.1):
        listaerr.append(errorcuad(drangonorm,curva_ms(drangonorm,i,j)[1]))
        listamusig.append([i,j])
```

```python
errmin=min(listaerr) #valor del error mínimo
pos=listaerr.index(errmin) #posicion del error minimo en la lista
print('El error minimo ',errmin, 'se encuentra en la posición ', pos)
musig=listamusig[pos] #valores de mu y sigma asociados a ese error
print('El valor de mu para obtener este error minimo es', musig[0], ' mientras que el de sigma es de ', musig[1])
```
